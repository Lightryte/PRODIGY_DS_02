```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_data = pd.read_csv('/content/test.csv')
train_data = pd.read_csv('/content/train.csv')
```

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

```python
train_data['Cabin_known'] = train_data['Cabin'].notna().astype(int)
train_data['Cabin_known'].value_counts()
```

```python
train_data.drop(['Cabin'], axis=1, inplace=True)
```

```python
train_data.isnull().sum()
```

```python
train_data.head(20)
```

```python
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
train_data.head(20)
```

```python
train_data['Family_Size'] = train_data['SibSp'] + train_data['Parch'] + 1
```

```python
train_data['Title'] = train_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train_data['Title'] = train_data['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')
```

```python
train_data['Title'] = train_data['Title'].replace('Mlle', 'Miss')
train_data['Title'] = train_data['Title'].replace('Ms', 'Miss')
train_data['Title'] = train_data['Title'].replace('Mme', 'Mrs')
```

```python
train_data = pd.get_dummies(train_data, columns=['Title'], drop_first=True)
```

```python
train_data.head(10)
```

```python
sns.countplot(x='Survived', data=train_data)
plt.title('Survival Distribution')
plt.figure(figsize=(5, 2))
plt.show()
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(5, 6))
sns.barplot(x='Pclass', y='Survived', hue='Sex_male', data=train_data)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=train_data)
plt.title('Age Distribution')
plt.show()
```
